```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
ds=pd.read_csv("customers.csv")
ds.head()
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
features=["Annual Income (k$)","Spending Score (1-100)"]
x=ds[features].copy()
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.35,min_samples=5)
labels=dbscan.fit_predict(x_scaled)
```

```python
ds['Cluster Labels']=labels
```

```python
no_clusters=len(set(labels))-(1 if -1 in labels else 0)
print(f"The number of clusters are:{no_clusters}")
no_noise=(labels==-1).sum()
print("Number of noise clusters are:",no_noise)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels)
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.title("DBSCAN of Income vs spending score")
plt.show()
```
